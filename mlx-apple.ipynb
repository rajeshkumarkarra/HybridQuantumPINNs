```python
# Lazy Evaluation
def fun1(x):
    return x + 1 # Dummy implementation
def expensive_fun(a):
    return a **2 

def fun(x):
    a = fun1(x)
    b = expensive_fun(a)
    return a, b
x = 5
y, _ = fun(x)
print("y = ", y)
```
